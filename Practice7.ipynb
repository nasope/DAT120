```python
def f(startValue, rent, years):
    increaseNumber = 1.0 + rent/100
    return startValue*(increaseNumber**years)
```

```python
year = [i for i in range(0,21)]
```

```python
a0 = 300000#kr
rent = 3#%

sums = [f(a0,rent,i) for i in year]
```

```python
import matplotlib.pyplot as plt

plt.plot(year,sums)
plt.title("300'000kr med 3% rente på 20 år")
plt.xlabel("År")
plt.ylabel("Kroner")
plt.show()
```

```python
def readFile() -> list:
    file = None
    name = ""
    while True:
        try:
            name = input("Skriv inn navn på filen: ")
            file = open(name, "r")
            break
        except:
            print("Navnet du skrev inn, finnes ikke")
            
    data = [line.split() for line in file]
    file.close()

    result = []
    
    for i in data:
        for j in i:
            if j == "":
                continue
            result.append(j)
    return result

data = readFile()
words = {}

for word in data:    
    if words.get(word) == None:
        words[word] = 1
        continue
    words[word] += 1

_ = [print(word+": "+str(words[word])) for word in words]
```
